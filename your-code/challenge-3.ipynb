```python
# Unique customers sorted by amount_spent:

unique_customers = orders.groupby('CustomerID')['amount_spent'].sum().reset_index()
unique_customers.head(10)
#unique_customers.shape
```

```python
#How to select customers whose aggregated amount_spent is in a given quantile range?

unique_customers['Tier'] = pd.qcut(unique_customers['amount_spent'], q = [0, .75, .95, 1], labels = ['normal tier', 'Preferred','VIP'])
# 0-75% , 75% - 99% , 99% - 100%
unique_customers
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
unique_customers.head()
#unique_customers.shape
```

```python
merged_orders = orders.merge(unique_customers, on = 'CustomerID')
#print(merged_orders)

#filtering only 3 columns:
merged_orders = merged_orders[['Country','CustomerID','Tier']]
#print(merged_orders)

# Only unique VIP customers:
condition = merged_orders['Tier'] == 'VIP'
merged_orders1 = merged_orders[condition].drop_duplicates('CustomerID')
#print(merged_orders1)

merged_orders2 = merged_orders1.groupby('Country')['CustomerID'].count().reset_index()
#print(merged_orders2)
merged_orders3 = merged_orders2.rename(columns = {'CustomerID': 'Number of VIP customers'})
merged_orders3.sort_values(by = 'Number of VIP customers' ,ascending = False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
#condition considering both VIP and preferred customers:
merged_orders = orders.merge(unique_customers, on = 'CustomerID')
condition = (merged_orders['Tier'] == 'VIP') | (merged_orders['Tier'] == 'Preferred')

#Only 3 columns:
merged_orders4 = merged_orders[condition][['Country','CustomerID','Tier']]
#print(merged_orders3)

#unique customers, and group by Country:
merged_orders5 = merged_orders4.drop_duplicates('CustomerID').groupby('Country')['CustomerID'].count().reset_index()
merged_orders6 = merged_orders5.rename(columns = {'CustomerID': 'Number of customers'})
merged_orders6.sort_values(by = 'Number of customers', ascending = False)
```
